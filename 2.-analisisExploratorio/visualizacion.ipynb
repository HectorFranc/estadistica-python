```python
import scipy
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../bicicletas-compartidas.csv')
```

## Categóricas
---

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: f'cuartil-{str(int(x))}')
```

**Histograma**

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Quarter')
ax.set_ylabel('Frecuency')
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Quarter')
ax.set_ylabel('Frecuency')

# Colorear
ax.patches[2].set_facecolor('red')
```

**Pie**

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels=y.value_counts().index)
ax.set_title('Pie diagram')
```

## Numericas
---

```python
y = df['viento']
```

**Histograma**

```python
fig, ax = plt.subplots()
ax.hist(y, bins=30, label='Viento') # bins es el número de particiones del histograma
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuency')
plt.axvline(np.mean(y), c='r', linestyle='--', label='Mean')
plt.axvline(np.mean(y) + np.std(y), c='k', linestyle='--', label='+ 1-std')
plt.axvline(np.mean(y) - np.std(y), c='k', linestyle='--', label='- 1-std')
ax.legend()
```

**Boxplot**

```python
y = df['bicis-compartidas']

sns.boxplot(x='cuartil-ano', y='bicis-compartidas', data=df)
```

**Scatterplot**

```python
fig, ax = plt.subplots()
ax.scatter(df['viento'], df['bicis-compartidas'], alpha=0.01)
ax.set_title('Bicis compartidas vs viento')
ax.set_xlabel('Viento')
ax.set_ylabel('Bicis compartidas')
```
